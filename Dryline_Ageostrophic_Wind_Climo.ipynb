```python
import xarray as xr
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np
import cartopy.crs as ccrs
import pandas as pd
import datetime as datetime

# path seagate
path_lr = 'E:/Honours_Data/'
print(path_lr)

UTC = ["0000","0100","0200","0300","0400","0500","0600","0700","0800","0900","1000","1100","1200","1300","1400","1500",
     "1600","1700","1800","1900","2000","2100","2200","2300"]
month = ["JAN","FEB","MAR","APR","MAY","JUN","JUL","AUG","SEP","OCT","NOV","DEC"]
monthlong = ["January","February","March","April","May","June","July","August","September","October","November","December"]
```

```python
control=xr.open_dataarray(path_lr+'Winds/u-ageo/April/uageoAPR0000mean.nc')
mthds_uageo=np.zeros((24,len(control.latitude),len(control.longitude)))
for y in range(0,12):
    for x in range(0,24):
        initial=xr.open_dataarray(path_lr+'Winds/u-ageo/'+monthlong[y]+'/uageo'+month[y]+UTC[x]+'mean.nc')
    #         initial=initial.assign_coords(time=x)
        mthds_uageo[x]=initial
# mthds_uageo
    uageo_mth=xr.DataArray(data=mthds_uageo,
                dims=['time','latitude','longitude'],
                coords=dict(
                time=UTC,
                latitude=control.latitude,
                longitude=control.longitude,
                ),
                attrs=dict(
                description=monthlong[y]+' Monthly and Hourly mean from 1990-2020 for the u-component of the ageostrophic wind',
                units='m/s')
                          )
    uageo_mth.to_netcdf(path=path_lr+'Winds/u-ageo/uageo'+month[y]+'hourlymean.nc')
                        
```

# start plotting and load data sets

```python
u=xr.open_dataarray(path_lr+'Winds/u-ageo/uageoFEBhourlymean.nc')
v=xr.open_dataarray(path_lr+'Winds/v-ageo/vageoFEBhourlymean.nc')
wsp=np.sqrt(u**2 + v**2)
condiv=xr.open_dataarray(path_lr+'Convergence/convFEBhourlymean.nc')
```

```python
# load data and filter large wind vectors
condiv=xr.open_dataset(path_lr+'Convergence/convANNUALhourlymean.nc')
ua=xr.open_dataset(path_lr+'Winds/u-ageo/uageoANNUALhourlymean.nc')
ua_n=ua.where(ua>-10)
ua_nan=ua_n.where(ua_n<10)
va=xr.open_dataset(path_lr+'Winds/v-ageo/vageoANNUALhourlymean.nc')
va_n=va.where(va>-10)
va_nan=va_n.where(va_n<10)
mslp=xr.open_dataset(path_lr+'MSLPANNUALhourlymean.nc')
# # #summer mean
summer_div=np.mean([condiv['conv_DEC'],condiv['conv_JAN'],condiv['conv_FEB']],axis=0)
summer_mslp=np.mean([mslp['mslp_DEC'],mslp['mslp_JAN'],mslp['mslp_FEB']],axis=0)
summer_ua=np.nanmean([ua_nan['ua_DEC'],ua_nan['ua_JAN'],ua_nan['ua_FEB']],axis=0)
summer_va=np.nanmean([va_nan['va_DEC'],va_nan['va_JAN'],va_nan['va_FEB']],axis=0)
summer_va[:,0]=np.nan
summer_ua[:,0]=np.nan

# spring mean
spring_div=np.mean([condiv['conv_SEP'],condiv['conv_OCT'],condiv['conv_NOV']],axis=0)
spring_mslp=np.mean([mslp['mslp_SEP'],mslp['mslp_OCT'],mslp['mslp_NOV']],axis=0)
spring_ua=np.nanmean([ua_nan['ua_SEP'],ua_nan['ua_OCT'],ua_nan['ua_NOV']],axis=0)
spring_va=np.nanmean([va_nan['va_SEP'],va_nan['va_OCT'],va_nan['va_NOV']],axis=0)
spring_va[:,0]=np.nan
spring_ua[:,0]=np.nan
```

```python
#Mean ageostrophic winds, convergence is the negative of diverence in this case
fig, ax = plt.subplots(2,2,subplot_kw={'projection': ccrs.PlateCarree()},figsize=(16,10))
time_AEST=["10am AEST","4pm AEST","10pm AEST","4am AEST"]
for x in range(0,1):
    time=(x*6)
    ax=ax.flatten()
    ax[x].set_extent([110, 155, -30, -10], ccrs.PlateCarree())
    lons, lats = np.meshgrid(ua.longitude, ua.latitude)
    lon_grid = np.arange(110,155,5)
    lat_grid = np.arange(-30,-10,5)
    gl = ax[x].gridlines(draw_labels=True,xlocs=lon_grid,ylocs=lat_grid,
              x_inline=False,y_inline=False,color='k',linestyle='--',linewidth=0.4)
    gl.top_labels = False
    gl.right_labels = False
    umean1=summer_ua[time]
    vmean1=summer_va[time]
    msl = ax[x].contour(lons,lats,summer_mslp[time],colors='k')
    kw_clabels = {'fontsize': 11, 'inline': True, 'inline_spacing': 5, 'fmt': '%i',
                      'rightside_up': True, 'use_clabeltext': True}
    ax[x].clabel(msl, **kw_clabels)
    conv = ax[x].contourf(lons,lats,summer_div[time], levels=[-1e-5,0,1e-5],extend='both',
                          colors=['tab:orange','white','white','tab:purple'])
    quiver_slices = (slice(None, None, 9), slice(None, None, 7))
    Q1=ax[x].quiver(lons[quiver_slices],lats[quiver_slices],(umean1[quiver_slices]),(vmean1[quiver_slices]),width=0.002,scale=130)
    kw_clabels = {'fontsize': 11, 'inline': True, 'inline_spacing': 5, 'fmt': '%i',
                      'rightside_up': True, 'use_clabeltext': True}
    ax[x].coastlines()
    ax[x].set_title(time_AEST[x])
    if x == 0:
        ax[0].quiverkey(Q1, 0.1, 0.92, 5, r'$5 \frac{m}{s}$', labelpos='E',
                           coordinates='figure',angle = 180, labelsep=0.3)
fig.suptitle('SUMMER Mean Ageostrophic Winds, MSLP and Divergence 1990-2020', fontsize='xx-large')
fig.subplots_adjust(bottom=0.3, top=0.9, left=0.1, right=0.9,
                    wspace=0.02, hspace=0.3)
    # # Add a big title at the top
#     plt.colorbar(conv, orientation='horizontal')
#     plt.title('FEB Mean Divergence, Geopotential Height and Ageostrophic Winds\n 1990-2020 at 925hPa',fontsize='xx-large')
#     plt.show()
```
